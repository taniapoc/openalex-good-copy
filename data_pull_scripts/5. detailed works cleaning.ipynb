```python
sfu_works_cleaned = sfu_works.drop(['ids', 'doi', 'title', 'publication_date', 'cited_by_percentile_year', 'biblio', 'indexed_in', 'apc_list', 'apc_paid', 'mesh', 
                                    'locations_count', 'locations', 'best_oa_location', 'grants', 'related_works', 'corresponding_author_ids', 'corresponding_institution_ids', 
                                    'abstract_inverted_index', 'concepts', 'topics', 'keywords', 'referenced_works', 'counts_by_year', 'updated_date', 'created_date'], 
                                    axis = 1)

# change url to regular openalex id
sfu_works_cleaned['open_alex_id'] = sfu_works_cleaned['id'].apply(lambda x: str(x).lstrip('https://openalex.org/'))

# chnge year to int
sfu_works_cleaned = sfu_works_cleaned[-sfu_works_cleaned['publication_year'].isna()]
sfu_works_cleaned['publication_year'] = sfu_works_cleaned['publication_year'].apply(lambda x: int(x))

# pull open access and status from open_access col
sfu_works_cleaned['is_open_access'] = sfu_works_cleaned['open_access'].apply(lambda x: ast.literal_eval(x)).apply(lambda y: y['is_oa'])
sfu_works_cleaned['open_access_status'] = sfu_works_cleaned['open_access'].apply(lambda x: ast.literal_eval(x)).apply(lambda y: y['oa_status'])
sfu_works_cleaned = sfu_works_cleaned.drop(['open_access'], axis=1)

# pull citation stats
sfu_works_cleaned['citation_percentile'] = sfu_works_cleaned['citation_normalized_percentile'].apply(lambda x: '' if str(x)=='nan' else ast.literal_eval(x)).apply(lambda y: np.nan if str(y)=='' else y['value'])
sfu_works_cleaned['top_1_pc_cited'] = sfu_works_cleaned['citation_normalized_percentile'].apply(lambda x: '' if str(x)=='nan' else ast.literal_eval(x)).apply(lambda y: np.nan if str(y)=='' else y['is_in_top_1_percent'])
sfu_works_cleaned['top_10_pc_cited'] = sfu_works_cleaned['citation_normalized_percentile'].apply(lambda x: '' if str(x)=='nan' else ast.literal_eval(x)).apply(lambda y: np.nan if str(y)=='' else y['is_in_top_10_percent'])
sfu_works_cleaned = sfu_works_cleaned.drop(['citation_normalized_percentile'], axis=1)

# pull topic, subfield, field, domain as before
sfu_works_cleaned['as_dict'] = sfu_works_cleaned['primary_topic'].apply(lambda x: '' if str(x)=='nan' else ast.literal_eval(x))
sfu_works_cleaned['topic'] = sfu_works_cleaned['as_dict'].apply(lambda x: np.nan if str(x)=='' else x['display_name'])
sfu_works_cleaned['subfield'] = sfu_works_cleaned['as_dict'].apply(lambda x: '' if str(x)=='' else x['subfield']).apply(lambda y: np.nan if str(y)=='' else y['display_name'])
sfu_works_cleaned['field'] = sfu_works_cleaned['as_dict'].apply(lambda x: '' if str(x)=='' else x['field']).apply(lambda y: np.nan if str(y)=='' else y['display_name'])
sfu_works_cleaned['domain'] = sfu_works_cleaned['as_dict'].apply(lambda x: '' if str(x)=='' else x['domain']).apply(lambda y: np.nan if str(y)=='' else y['display_name'])
sfu_works_cleaned = sfu_works_cleaned.drop(['primary_topic', 'as_dict'], axis=1)

# extract publication location
sfu_works_cleaned['location_dict'] = sfu_works_cleaned['primary_location'].apply(lambda x: '' if str(x)=='nan' else ast.literal_eval(x))
sfu_works_cleaned['source_dict'] = sfu_works_cleaned['location_dict'].apply(lambda x: np.nan if str(x)=='' else x['source'])
sfu_works_cleaned['source_dict'] = sfu_works_cleaned['source_dict'].apply(lambda x: np.nan if str(x)=='None' else x)
sfu_works_cleaned['Source Name'] = sfu_works_cleaned['source_dict'].apply(lambda x: '' if str(x)=='nan' else x['display_name'])
sfu_works_cleaned['Source Type'] = sfu_works_cleaned['source_dict'].apply(lambda x: '' if str(x)=='nan' else x['type'])

# pull publication SDGs
    # NOTE: openalex stores only SDGs with a prediction score (using some algorithm) greater than 0.4, but all of the publications so far only have 0, 1, or 2 SDGs associated, 
    # so I will only be storing the top SDG for each publicatio, if it exists 
sfu_works_cleaned['SDG_number'] = sfu_works_cleaned['sustainable_development_goals'].apply(lambda x: np.nan if str(x)=='[]' else ast.literal_eval(x)[0]['id'].lstrip('https://metadata.un.org/sdg/'))
sfu_works_cleaned['SDG_name'] = sfu_works_cleaned['sustainable_development_goals'].apply(lambda x: np.nan if str(x)=='[]' else ast.literal_eval(x)[0]['display_name'])
sfu_works_cleaned = sfu_works_cleaned.drop(['sustainable_development_goals'], axis=1)

# get collaboration info
sfu_works_cleaned['Collaboration Status'] = ''
conditions = [
    ((sfu_works_cleaned['Collab. Countries'] == 1) & (sfu_works_cleaned['Collab. Institutions'] == 1)), 
    ((sfu_works_cleaned['Collab. Countries'] == 1) & (sfu_works_cleaned['Collab. Institutions'] > 1)), 
    (sfu_works_cleaned['Collab. Countries'] > 1)
]
choices = ['SFU Only', 'Canadian Collab.', 'Int\'l Collab.']
sfu_works_cleaned['Collaboration Status'] = np.select(conditions, choices, default='')


# deal with authorship data
    # NOTE: this data is pretty difficult to organize nicely, since some papers have only 1 author whereas others have 100+ (openalex truncates the list at 100 for the sake of storage I assume)
sfu_works_cleaned['authors_json'] = sfu_works_cleaned['authorships'].apply(lambda x: ast.literal_eval(x))
sfu_works_cleaned['author_count'] = sfu_works_cleaned['authors_json'].apply(lambda x: len(x)).apply(lambda y: '100+' if str(y)=='100' else y)
sfu_works_cleaned['first_author_detail'] = sfu_works_cleaned['authors_json'].apply(lambda x: x[0])
sfu_works_cleaned['citation_percentile'] = sfu_works_cleaned['citation_percentile'].apply(lambda x: x*100)
sfu_works_cleaned['first_author_name'] = sfu_works_cleaned['first_author_detail'].apply(lambda x: x['author']['display_name'])
sfu_works_cleaned['first_author_sfu'] = sfu_works_cleaned['first_author_detail'].apply(lambda x: x['institutions']).apply(lambda y: str(y).find('Simon Fraser University')).apply(lambda z: False if z==-1 else True)
sfu_works_cleaned = sfu_works_cleaned.drop(['authorships', 'authors_json', 'first_author_detail', 'id'], axis=1)
sfu_works_cleaned = sfu_works_cleaned.rename(columns={'display_name': 'Title', 'publication_year': 'Year', 'language': 'Language', 'type': 'Publication Type', 'countries_distinct_count': 'Collab. Countries', 
                                                      'institutions_distinct_count': 'Collab. Institutions', 'fwci': 'FWCI', 'cited_by_count': 'Citations Received', 'is_retracted': 'Is Retracted', 'is_paratext': 'Is Paratext', 
                                                      'referenced_works_count': 'Works Referenced', 'open_alex_id': 'OpenAlex ID', 'is_open_access': 'Is Open Access', 'open_access_status': 'Open Access Status', 
                                                      'citation_percentile': 'Citation Percentile', 'top_1_pc_cited': 'Top 1% Cited', 'top_10_pc_cited': 'Top 10% Cited', 'topic': 'Topic', 'subfield': 'Subfield', 'field': 'Field', 
                                                      'domain': 'Domain', 'SDG_number': 'Top Matching SDG', 'SDG_name': 'SDG Name', 'author_count': 'Author Count', 'first_author_name': 'First Author', 'first_author_sfu': 'First Author From SFU'})
sfu_works_cleaned = sfu_works_cleaned[['OpenAlex ID', 'Title', 'Year', 'FWCI', 'Citations Received', 'Citation Percentile', 'Top 1% Cited', 'Top 10% Cited', 'Top Matching SDG', 'SDG Name', 'Topic', 'Subfield', 'Field', 'Domain', 
                                       'Language', 'Publication Type', 'Is Retracted', 'Is Paratext', 'Works Referenced', 'Collab. Countries', 'Collab. Institutions', 'Author Count', 'First Author', 'First Author From SFU', 
                                       'Is Open Access', 'Open Access Status', 'source_dict', 'Source Name', 'Source Type', 'Collaboration Status']]

# print
sfu_works_cleaned.head()
```

### BONUS: fill in summary stats

```python
current_yr = datetime.now().year

summary = pd.DataFrame(pd.read_csv('../data_pulls/sfu_summary_stats.csv'))

# calculate 5 and 10 yr publication and citation counts
sfu_publications_5yr = np.sum(sfu_counts_by_year[sfu_counts_by_year['Year'] > (current_yr-5)]['Publications'])
sfu_publications_10yr = np.sum(sfu_counts_by_year[sfu_counts_by_year['Year'] > (current_yr-10)]['Publications'])

sfu_citations_5yr = np.sum(sfu_counts_by_year[sfu_counts_by_year['Year'] > (current_yr-5)]['Citations Received'])
sfu_citations_10yr = np.sum(sfu_counts_by_year[sfu_counts_by_year['Year'] > (current_yr-10)]['Citations Received'])

# import counts into summary table
summary.loc[summary['Time Frame']=='5 years', 'Publications'] = sfu_publications_5yr
summary.loc[summary['Time Frame']=='10 years', 'Publications'] = sfu_publications_10yr

summary.loc[summary['Time Frame']=='5 years', 'Citations'] = sfu_citations_5yr
summary.loc[summary['Time Frame']=='10 years', 'Citations'] = sfu_citations_10yr

# calculate citations per publication
summary['Citations per Publication'] = summary['Citations']/summary['Publications']

summary

summary.to_csv('../data_pulls/sfu_summary_stats.csv', index=False)
```
